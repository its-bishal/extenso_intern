```python
injuries_spdcaseno_df=pd.crosstab(df['SPDCASENO'], df['INJURIES'])
```

```python
from scipy.stats import chi2_contingency

# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(injuries_spdcaseno_df)


# Print results
print("Chi-Square Statistic:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)
```

```python
from scipy.stats import shapiro

data = df[['PERSONCOUNT','PEDCOUNT', 'PEDCYLCOUNT', 'VEHCOUNT', 'INJURIES', 'SERIOUSINJURIES', 'FATALITIES']]
statistic, p_value = shapiro(data)

print("Shapiro-Wilk Test Statistic:", statistic)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05  # significance level
if p_value > alpha:
    print("The data looks normally distributed (fail to reject H0)")
else:
    print("The data does not look normally distributed (reject H0)")
```

This is an indication that the data is not normally distributed accross all its range.

```python
from scipy.stats import pearsonr

# Perform Pearson correlation test
corr_coefficient, p_value = pearsonr(df['PERSONCOUNT'], df['INJURIES'])

# Print results
print("Pearson Correlation Coefficient:", corr_coefficient)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05  # significance level
if p_value < alpha:
    print("There is a significant linear relationship between the variables (reject H0)")
else:
    print("There is no significant linear relationship between the variables (fail to reject H0)")
```

The result from Pearson Correlation Coefficient indicates that the count of injuries is higher  in the dense locations than the other places, and the scatter plot plotted in between VEHCOUNT, INJURIES and SDOT_COLCODE, indicates, there is no proportional rise between injuries  and increase in vehicle counts.
