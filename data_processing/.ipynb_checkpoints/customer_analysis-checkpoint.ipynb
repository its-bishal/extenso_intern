###  Perform monthly and weekly level product wise aggregations of the transactions for each user 

```python
trans_data.head()
```

```python
trans_data.dtypes
```

```python
trans_data['last_modified_date'] = pd.to_datetime(trans_data['last_modified_date'])
```

```python
grouped = trans_data.groupby(['payer_account_id', 'product_id', 'product_type_id', pd.Grouper(key='last_modified_date', freq='M')])
monthly_agg = grouped['amount'].mean().reset_index()
```

```python
monthly_agg
```

```python
monthly_agg['payer_account_id'].nunique()
```

```python
monthly_agg1 = trans_data.groupby('payer_account_id').agg({'amount': 'mean'}).reset_index()
monthly_agg1
```

```python
monthly_agg2 = (
    trans_data.groupby(['payer_account_id', pd.Grouper(key='last_modified_date', freq='M')])
    .agg({'amount': 'mean', 'product_id': 'first', 'product_type_id': 'first'})
    .reset_index()
)
monthly_agg2
```

```python
grouped1 = trans_data.groupby(['payer_account_id' , 'product_id', 'product_type_id', pd.Grouper(key='last_modified_date', freq='W-MON')])
weekly_agg = grouped1['amount'].mean().reset_index()
```

```python
weekly_agg
```

```python
weekly_agg1 = weekly_agg.groupby('payer_account_id').agg({'amount': 'mean'}).reset_index()
weekly_agg1
```

```python
merged_df = pd.merge(monthly_agg1, weekly_agg1, on='payer_account_id', how='outer')
merged_df
```

```python
merged_df = merged_df.rename(columns={'amount_x':'monthly_avg'})
merged_df = merged_df.rename(columns={'amount_y':'weekly_avg'})
merged_df
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

# Fit a Gaussian distribution to the data
mu, std = norm.fit(trans_data['amount'])

# Plot the histogram of the data
plt.hist(trans_data['amount'], bins=5, density=True, alpha=0.6, color='b')

# Plot the PDF (Probability Density Function) of the fitted Gaussian distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 10)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Gaussian Distribution Fit')

# Show the plot
plt.show()
```

```python
rew_p = trans_data.groupby('payer_account_id').agg({'reward_point': 'sum'})
merged_df = pd.merge(merged_df, rew_p, on='payer_account_id', how='outer')
```

```python
cash_b = trans_data.groupby('payer_account_id').agg({'cash_back_amount':'sum'})
merged_df = pd.merge(merged_df, cash_b, on='payer_account_id', how='outer')
```

```python
rev_amt = trans_data.groupby('payer_account_id').agg({'revenue_amount':'sum'})
merged_df = pd.merge(merged_df, rev_amt, on='payer_account_id', how='outer')
merged_df
```

```python
total_data = pd.merge(prod_cat_map, trans_data, on=['module_id','product_id','product_type_id'], how='inner')
total_data
```

```python
new = total_data.groupby(['payer_account_id','txn_flow']).agg({'amount': 'sum'}).reset_index()
new.columns
```

```python
# Convert categorical column to multiple binary columns
dummy_df = pd.get_dummies(new['txn_flow'])

# Concatenate the dummy columns with the original DataFrame
result_df = pd.concat([new, dummy_df], axis=1)
result_df
```

```python
pivot_df = total_data.pivot_table(index='payer_account_id', columns='txn_flow', values='amount', aggfunc='sum')
pivot_df
```

```python
merged_df = pd.merge(merged_df, pivot_df, on='payer_account_id', how='inner')
merged_df
```
