```python
import pandas as pd
from kafka import KafkaProducer
from datetime import datetime
import time
import random
import numpy as np

```

```python
# 2nd topic
KAFKA_TOPIC_NAME_CONS2 = "water_portability"
KAFKA_SERVER_BOOTSTRAP_CONS2 = "localhost:9092"

if __name__ == "__main__":
    print("kafka producer application started...")

    kafka_producerObj = KafkaProducer(bootstrap_servers=KAFKA_SERVER_BOOTSTRAP_CONS2,
                                      value_serializer=lambda x: x.encode('utf-8'))

    filepath1 = "B:/data/water_quality/water_potability.csv"

    water_df = pd.read_csv(filepath1)

    water_list = water_df.to_dict(orient='records')

    message_list = []
    message = None

    for message in water_list:
        msg_fields_value_list=[]
        
        # message_fields_value_list.append(message["order_id"])
        msg_fields_value_list.append(message["ph"])
        msg_fields_value_list.append(message["Hardness"])
        msg_fields_value_list.append(message["Solids"])
        msg_fields_value_list.append(message["Chloramines"])
        msg_fields_value_list.append(message["Sulfate"])
        msg_fields_value_list.append(message["Conductivity"])
        msg_fields_value_list.append(message["Organic_carbon"])
        msg_fields_value_list.append(message["Trihalomethanes"])
        msg_fields_value_list.append(message["Potability"])


        message = ','.join(str(v) for v in msg_fields_value_list)
        print("message tyoe: ",type(message))

        print("Message: ", message)
        kafka_producerObj.send(KAFKA_TOPIC_NAME_CONS2, message)
        time.sleep(1)

    print("Kafka Producer Application Completed.")
              
```
